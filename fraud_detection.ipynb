```python
print(data.columns)  # Check if 'isFraud' exists
print(data.dtypes)   # Check column types
```

```python
# Check if column exists
if 'isFraud' in data.columns:
    # Convert categorical columns to numeric
    data_numeric = data.copy()
    for col in data_numeric.select_dtypes(include=['object']).columns:
        data_numeric[col] = pd.factorize(data_numeric[col])[0]
    
    correlation = data_numeric.corr()
    print(correlation['isFraud'].sort_values(ascending=False))
else:
    print("Error: 'isFraud' column not found in data. Available columns:", data.columns.tolist())
```

```python
#convert the categorical values into numreical values
data['type']=data['type'].map({'CASH_OUT':1,'PAYMENT':2,
                              'CASH_IN':3,'TRANSFER':4,
                              'DEBIT':5})
data['isFraud']=data['isFraud'].map({0:'No Fraud',1:'Fraud'})
print(data.head())
```

```python
from sklearn.model_selection import train_test_split
x=np.array(data[['type','amount','oldbalanceOrg','newbalanceOrig',]])
y=np.array(data[['isFraud']])
```

```python
#traning a machine learning model
from sklearn.tree import DecisionTreeClassifier
xtrain , xtest , ytrain , ytest = train_test_split(
    x,y,test_size=0.10,random_state=42)
model=DecisionTreeClassifier()
model.fit(xtrain,ytrain)
model.score(xtest,ytest)
```

```python
#prediction
#features=['type','amount','oldbalanceOrg','newbalanceOrig','oldbalanceDest','newbalanceDest']
featurees=np.array([[4,181.00,181.0,0.0,0.0,0.0]])

print(model.predict(featurees))
```
